```python
import pandas as pd
```

```python
metro_col = pd.read_csv('/home/shumway743/cis5500_group/cost_of_living/metro_col.csv')
metro_col.head()
```

```python
col = pd.read_csv('/home/shumway743/cis5500_group/cost_of_living/combined_cost_of_living.csv')
col.head()
```

```python
col['area_name_lower'] = col.areaname.str.lower()
col.head()
```

```python
jdf = census_county.merge(col, left_on='area_name', right_on='area_name_lower', how='inner')
jdf = jdf[['GEO_ID','annual_cost','index']]
county_out = jdf.rename({'index':'cost_index'}, axis=1)
county_out.head()
```

```python
jdf.count()
```

```python
jdf[jdf.area_name_lower.isnull()].head()
```

```python
def get_match_metro(x):
    m = difflib.get_close_matches(x['areaname'], census_metro[x['state'] == census_metro['state_code']].NAME,n=1, cutoff=.60)
    return m[0] if len(m) else None

jdf_metro = col[col.type == 'metro']
jdf_metro['match'] = jdf_metro.apply(get_match_metro, axis=1)
```

```python
jdf_metro_sel = jdf_metro[['annual_cost','index','state','match']]
jdf_metro_sel.head()
```

```python
metro_out = census_metro.merge(jdf_metro_sel, left_on='NAME', right_on='match', how='inner')
metro_out = metro_out[['GEO_ID','annual_cost','index']].rename({'index':'cost_index'}, axis=1)
metro_out.head()
```

```python
census_metro.count()
```

```python
combined = pd.concat([county_out, metro_out])
combined.head()
```
