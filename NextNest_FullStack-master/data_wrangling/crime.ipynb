```python
import difflib
census_city = census[census.SUMLEVEL.isin({160})]
census_city['area_name'] = census_city.NAME.apply(lambda x: x.split(', ')[0])
census_metro = census[census.SUMLEVEL.isin({310})]
census_metro['state_code'] = census_metro.NAME.apply(lambda x: x.split(', ')[1].split(' ')[0])
census_metro['area_name'] = census_metro.NAME.apply(lambda x: x.split(', ')[0])
census_metro.head()
def get_match_metro1(x):
    m = difflib.get_close_matches(x, city_3.name_state_abv,n=1, cutoff=.75)
    return m[0] if len(m) else None
def get_match_city1(x):
    m = difflib.get_close_matches(x['NAME'], city_3[city_3.abv == x['STUSAB']].name_state,n=1, cutoff=.75)
    return m[0] if len(m) else None

def get_match_metro(x):
    m = difflib.get_close_matches(x['area_name'].lower(), census_metro[x['abv'] == census_metro['state_code']].area_name.str.lower(),n=1, cutoff=.6)
    return m[0] if len(m) else None
def get_match_city(x):
    m = difflib.get_close_matches(x['clean_name'].lower(), census_city[x['abv'] == census_city['STUSAB']].clean_name.str.lower(),n=1, cutoff=.6)
    return m[0] if len(m) else None
```

```python
import re
words = {}
def word_count(x):
    s = re.sub("[^\w' ]", "", x.lower()).split()
    for w in s:
        words[w] = words.get(w,0) + 1
census_city.area_name.apply(word_count)
city_3.area_name.apply(word_count)

# census_city.clean_name.apply(word_count)
# city_3.clean_name.apply(word_count)




drop_words = set([x[0] for x in list(filter(lambda x: x[1] > 1000, words.items()))])
print(drop_words)

def remove_words(x):
    out = x
    for w in list(drop_words):
        out = out.lower().replace(w,'')
    return out

census_city['clean_name'] = census_city.area_name.apply(remove_words)
city_3['clean_name'] = city_3.area_name.apply(remove_words)
# print(words)
```
